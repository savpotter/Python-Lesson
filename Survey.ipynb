```python
import pandas as pd

def python_and_excel_quiz():
    print("Welcome to the Python and Excel Survey!\n")

    questions = [
        {
            "question": "How would you rate your understanding of Python before this lesson?",
            "options": ["A. Very confident", "B. Somewhat confident", "C. Neutral", "D. Not confident", "E. No understanding at all"]
        },
        {
            "question": "How would you rate your understanding of Excel before this lesson?",
            "options": ["A. Very confident", "B. Somewhat confident", "C. Neutral", "D. Not confident", "E. No understanding at all"]
        },
        {
            "question": "How confident do you feel about using Python after this lesson?",
            "options": ["A. Very confident", "B. Somewhat confident", "C. Neutral", "D. Not confident", "E. Not at all"]
        },
        {
            "question": "How confident do you feel about using Excel after this lesson?",
            "options": ["A. Very confident", "B. Somewhat confident", "C. Neutral", "D. Not confident", "E. Not at all"]
        },
        {
            "question": "How well was the material presented during the lesson?",
            "options": ["A. Extremely well", "B. Very well", "C. Adequately", "D. Poorly", "E. Very poorly"]
        },
        {
            "question": "What was the most challenging part of this lesson for you?",
            "options": []
        },
        {
            "question": "What was the most helpful part of this lesson for you?",
            "options": []
        },
        {
            "question": "What suggestions do you have for improving this lesson?",
            "options": []
        }
    ]

    responses = []

    for i, q in enumerate(questions):
        print(f"Question {i + 1}: {q['question']}")
        if q['options']:
            for option in q['options']:
                print(option)
            user_answer = input("Your answer (A/B/C/D/E): ").strip().upper()
        else:
            user_answer = input("Your response: ").strip()
        responses.append(user_answer)

    feedback_data = {
        "Question": [q['question'] for q in questions],
        "Response": responses
    }

    feedback_df = pd.DataFrame(feedback_data)
    feedback_file = "quiz_feedback.xlsx"

    feedback_df.to_excel(feedback_file, index=False)
    print(f"Feedback saved to {feedback_file}. Thank you for your responses!")

if __name__ == "__main__":
    python_and_excel_quiz()
```
